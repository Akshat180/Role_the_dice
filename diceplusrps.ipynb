```python
def rps(diceWinner):
    print("rock paper scissor Game: ")
    print("")
    global c1, c2
    c1 = 0
    c2 = 0
    for i in range(1,3):
        if diceWinner:
            user1choice = input('user 1: ')
            user2choice = input('user 2: ')
        else:
            user2choice = input('user 2: ')
            user1choice = input('user 1: ')
        if user1choice == user2choice:
            c1+=1
            c2+=1
            print("Match is Draw ")
        elif (user1choice == "paper" and user2choice == "rock") or (user1choice == "rock" and user2choice == "scissor") or (user1choice == "scissor" and user2choice == "paper"):
            c1+=1
            print("User 1 win: ")
        else:
            c2+=1
            print("User 2 win: ")
            print("")
        
    if c1>c2:
        print("User 1 win by: ",c1)
    elif c2>c1:
        print("User 2 win by: ",c2)
    else:
        print('Game is Draw')
```

```python
def dice():
    import random
    k = random.choice(range(1,7))
    if(k==1):
        print('[-----]')
        print('[-----]')
        print('[--0--]')
        print('[-----]')
        print('[-----]')
    elif(k==2):
        print('[-----]')
        print('[0    ]')
        print('[-----]')
        print('[    0]')
        print('[-----]')
    elif(k==3):
        print('[-----]')
        print('[     ]')
        print('[0 0 0]')
        print('[     ]')
        print('[-----]')
    elif(k==4):
        print('[-----]')
        print('[0   0]')
        print('[     ]')
        print('[0   0]')
        print('[-----]')
    elif(k==5):
        print('[-----]')
        print('[0   0]')
        print('[  0  ]')
        print('[0   0]')
        print('[-----]')
    elif(k==6):
        print('[-----]')
        print('[0 0 0]')
        print('[     ]')
        print('[0 0 0]')
        print('[-----]')
    print('inside', k)
    return k
```

```python
print('Want to play the game')

n = input("Y/N: ").upper()
if n=="Y":
    flag = 1
    while flag:
        user1 = dice()
        user2 = dice()
        if user1 > user2:
            flag = 0            
            print("rock","paper","scissor","\n")
            rps(1)

        elif user2 > user1: 
            flag = 0
            print("rock","paper","scissor","\n")
            rps(0)

        elif user1 == user2:
            print('Same dice. Draw. Roll again')
            flag = 1


        else:
            print("none")
```
